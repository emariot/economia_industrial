# Razão de Concentração
### A Razão de Concentração é importante para determinar a participação de grandes empresas no mercado. Ele varia de 0 a 1 onde 1 indica uma situação de concentração instensa.
* CR(4)= 0.98 O índice para as 4 maiores empresas indica um nível de concentração muito alto.  
* CR(8)= 0.99 O índice para as 8 maiores empresas indica um nível de concentração muito alto.  


# CR(4)

```python
CR_4 = df_aux.total[0:4].sum()
print(f'{CR_4:.2f} %')
```

# CR(8)

```python
CR_8 = df_aux.total[0:8].sum()
print(f'{CR_8:.2f}%')
```

# Índice de Herfindal - Hirschman (HH)
### É um índice que mensura a concentração do mercado e da concorrência entre os participantes, porque considera o total das empresas e não apenas as maiores.
### Quanto maior for HH, mais elevada será a concentração e menor a concorrência entre as empresas.


```python
df_aux
```

```python
h = round(df_aux.total*100,2)
h
```

```python
hh = 0
for i in range(len(h)):
    p = (h[i]**2)
    hh += p
 
```

```python
print(f'indice HH: {hh:.0f}')
```

```python
df_aux_novo.reset_index(drop=True, inplace=True)
df_aux_novo
```

```python
hn = round(df_aux_novo*100,2)
hn
```

```python
hhn = 0
for i in range(len(hn)):
    p = (hn[i]**2)
    hhn += p
```

```python
print(f'indice HH: {hhn:.0f}')
```
